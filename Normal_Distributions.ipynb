```python
import numpy as np
import scipy.stats as stats

# Parameters for the normal distribution
mu = 75  # mean grade
sigma = 10  # standard deviation

# Calculate the CDF for 85 and 65
p_85 = stats.norm.cdf(85, mu, sigma)
p_65 = stats.norm.cdf(65, mu, sigma)

# Probability of scoring above 85
prob_above_85 = 1 - p_85

# Probability of scoring below 65
prob_below_65 = p_65

# Print the results
print(f"Percentage of students scoring above 85: {prob_above_85 * 100:.2f}%")
print(f"Percentage of students scoring below 65: {prob_below_65 * 100:.2f}%")
```

```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mu = 75  # mean grade
sigma = 10  # standard deviation

# Create an array of grades
x = np.linspace(50, 100, 1000)

# Calculate the normal distribution (PDF)
pdf = stats.norm.pdf(x, mu, sigma)

# Calculate the CDF for thresholds
p_85 = stats.norm.cdf(85, mu, sigma)
p_65 = stats.norm.cdf(65, mu, sigma)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# Fill areas under the curve for probabilities
plt.fill_between(x, pdf, where=(x >= 85), color='orange', alpha=0.5, label='P(X > 85)')
plt.fill_between(x, pdf, where=(x <= 65), color='red', alpha=0.5, label='P(X < 65)')

# Add vertical lines for thresholds
plt.axvline(85, color='orange', linestyle='--', label='Score = 85')
plt.axvline(65, color='red', linestyle='--', label='Score = 65')

# Add titles and labels
plt.title('Distribution of Machine Learning Grades at ALU')
plt.xlabel('Grades')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()

# Show the plot
plt.show()


```
